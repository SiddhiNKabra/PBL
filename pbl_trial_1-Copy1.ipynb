```python
rain_data=tc.SFrame("rainfall.csv")
```

```python
rain_feature=["SUBDIVISION","YEAR","JAN","FEB","MAR","APR","MAY","JUN","JUL","AUG","SEP","OCT","NOV","DEC"]
```

```python
rainfall_model=tc.linear_regression.create(rain_train,target="total",features=rain_feature,validation_set=None)
```

```python
metrics=rainfall_model.evaluate(rain_test)
```
